# 线性回归预测

## 1、线性回归介绍
线性回归是一种用于建立自变量（输入特征）和因变量（输出标签）之间关系的模型。它是最简单的机器学习算法之一，广泛应用于预测
和回归任务。
线性回归模型可以表示为：  
$$y=\beta_0+\beta_1x+\epsilon$$

$$\text{其中, y是因变量, x是自变量, }\beta_0\text{ 是截距, }\beta_1\text{是斜率, }\epsilon\text{ 而是误差项。}$$

### 2.1 导入所需库
在构建模型之前，需要先导入模型所需要的库。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pandas as pd
```

### 2.2 读取CSV文件
假设我们有一个名为data.csv的CSV文件，其中包含两列数据：X和y。

```python
# 读取CSV文件
data = pd.read_csv('../数据集/data.csv')
# 提取特征和标签
X = pd.DataFrame(data['X'])
y = data['y']
```

### 2.3 数据预处理
在训练模型之前，我们需要对数据进行预处理。将数据划分为训练集和测试集。

```python
# 将数据集分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)
```

### 2.4 创建线性回归模型
现在，我们可以创建一个线性回归模型。


```python
# 创建线性回归模型
model = LinearRegression()
```

### 2.5 训练模型
接下来，我们将使用训练数据集来训练模型。


```python
# 训练模型
model.fit(X_train, y_train)
```

### 2.6 评估模型
训练完成后，我们可以使用测试数据集来评估模型的性能。

```python
# 预测测试集
y_pred = model.predict(X_test)

# 模型评分
print("模型准确率（满分1）：{}".format(r2_score(y_test,y_pred)))
print("模型得到如下计算公式：")
print("y = X ✖ {} + {}".format(model.coef_[0], model.intercept_))
```

### 2.7 可视化
最后，我们可以将数据集和模型的预测结果可视化。


```python
# 可视化
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

这将显示实际数据点和通过线性回归模型获得的预测数据点。
